```python
from sympy import *

a = float(0.6)
b = float(5.5)
c = 10
d = -5

x = Symbol("x")

f_x = a * (x ** 3) + b * (x ** 2) + c * (x ** 1) + d
```

```python
f_x
```

```python
print("корень 1: ", solve(f_x,x)[0].n())
print("корень 2: ", solve(f_x,x)[1].n())
print("корень 3: ", solve(f_x,x)[2].n())
```

```python
from sympy.plotting import plot

q1 = plot(f_x,(x, -8,8), line_color= 'blue', title = 'График функций')
```

```python
der = diff(f_x)
der
```

```python
q1 = plot(der,(x, -8,8), line_color= 'blue', title = 'График функций')
```

```python
q2 = plot(f_x, der,(x, -8,8), line_color= 'blue', title = 'Совмещенный график' , leg=True,xlab = 'X', ylab = 'Y',show=False )

q2[0].line_color = 'blue'
q2[1].line_color = 'yellow'

q2.show()
```

```python
print("корень 1: ", solve(der,x)[0].n(3))
print("корень 2: ", solve(der,x)[1].n(3))

```

```python
solve(der > 0 , x)
```

```python
solve(der < 0 , x)
```
